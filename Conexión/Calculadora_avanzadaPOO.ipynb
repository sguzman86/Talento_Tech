```python
class CalculadoraAvanzada:
    def __init__(self):
        self.mostrar_menu()

    def mostrar_menu(self):
        print("\nBienvenido a la calculadora avanzada. Operaciones disponibles:")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicación")
        print("4. División")
        print("5. Potencia")
        print("6. Raíz cuadrada")
        print("7. Factorial")
        print("8. Salir")
        self.obtener_operacion()

    def obtener_operacion(self):
        opcion = input("\nIngrese el número de la operación: ")

        if opcion == "8":
            print("\nGracias por usar la calculadora. ¡Hasta pronto!")
            return

        if opcion in ["1", "2", "3", "4", "5"]:
            self.obtener_numeros(opcion)
        elif opcion in ["6", "7"]:
            self.obtener_un_numero(opcion)
        else:
            print("\nOpción no válida. Intente nuevamente.")
            self.mostrar_menu()

    def obtener_numeros(self, opcion):
        try:
            num1 = float(input("Ingrese el primer número: "))
            num2 = float(input("Ingrese el segundo número: "))
            self.realizar_operacion(opcion, num1, num2)
        except ValueError:
            print("\nError: Ingrese números válidos.")
            self.mostrar_menu()

    def obtener_un_numero(self, opcion):
        try:
            num = float(input("Ingrese el número: "))
            self.realizar_operacion_unaria(opcion, num)
        except ValueError:
            print("\nError: Ingrese un número válido.")
            self.mostrar_menu()

    def realizar_operacion(self, opcion, num1, num2):
        if opcion == "1":
            resultado = num1 + num2
            print(f"\nResultado: {num1} + {num2} = {resultado}")
        elif opcion == "2":
            resultado = num1 - num2
            print(f"\nResultado: {num1} - {num2} = {resultado}")
        elif opcion == "3":
            resultado = num1 * num2
            print(f"\nResultado: {num1} * {num2} = {resultado}")
        elif opcion == "4":
            if num2 == 0:
                print("\nError: No se puede dividir por cero.")
            else:
                resultado = num1 / num2
                print(f"\nResultado: {num1} / {num2} = {resultado}")
        elif opcion == "5":
            resultado = num1 ** num2
            print(f"\nResultado: {num1} ^ {num2} = {resultado}")

        self.mostrar_menu()

    def realizar_operacion_unaria(self, opcion, num):
        import math

        if opcion == "6":
            if num < 0:
                print("\nError: No existe raíz de número negativo.")
            else:
                resultado = math.sqrt(num)
                print(f"\nResultado: √{num} = {resultado}")
        elif opcion == "7":
            if num < 0:
                print("\nError: No existe factorial de número negativo.")
            elif not num.is_integer():
                print("\nError: El factorial requiere número entero.")
            else:
                resultado = math.factorial(int(num))
                print(f"\nResultado: {int(num)}! = {resultado}")

        self.mostrar_menu()


# Iniciar la calculadora
if __name__ == "__main__":
    print("\nIniciando Calculadora Avanzada...")
    calculadora = CalculadoraAvanzada()
```
