**Compilación del código del ejercicio de tablas**

```python
import pandas as pd

# Crear listas de datos
names = ["Andrew", "Phillip", "John", "Felipe", "Gabriela", "Luis", "Carlos", "Juliana", "Pedro", "Rosa"]
weights = [50, 78, 91.5, 58.5, 77, 68.5, 66, 90.5, 80, 86]  # peso en kilogramos
heights = [1.75, 1.87, 1.69, 1.57, 1.92, 1.78, 1.82, 1.67, 1.69, 1.79]  # altura en metros

# Crear DataFrame
df = pd.DataFrame({"name": names, "weight": weights, "height": heights})

# Calcular índice de masa corporal (IMC)
df["bmi"] = df["weight"] / df["height"]**2

# Clasificar el IMC
def clasificar_imc(imc):
    if imc < 18.5:
        return "Bajo peso"
    elif imc < 25:
        return "Promedio"
    else:
        return "Sobrepeso"

df["clasificación"] = df["bmi"].apply(clasificar_imc)

# Mostrar DataFrame final
print(df)
df
```
