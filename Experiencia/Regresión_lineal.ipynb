```python
import pandas as pd
import numpy as np

# Paso 1: Crear el DataFrame con los datos originales
data = {
    'Y': [78.5, 74.3, 104.3, 87.6, 95.9, 109.2, 102.7, 72.5, 93.1, 115.9, 83.8, 113.3, 109.4],
    'X1': [7, 1, 11, 11, 7, 11, 3, 1, 2, 21, 1, 11, 10],
    'X2': [26, 29, 56, 31, 52, 55, 71, 31, 54, 47, 40, 66, 68],
    'X3': [6, 15, 8, 8, 6, 9, 17, 22, 18, 4, 23, 9, 8],
    'X4': [60, 52, 20, 47, 33, 22, 6, 44, 22, 26, 34, 12, 12]
}
df = pd.DataFrame(data)

# Paso 2: Crear la matriz X con columna de unos (intercepto)
X = df[['X1', 'X2', 'X3', 'X4']]
X.insert(0, 'Intercepto', 1)

# Paso 3: Variable dependiente Y
Y = df[['Y']]

# Paso 4: Convertir a matrices NumPy
X_matrix = X.values
Y_matrix = Y.values

# Paso 5: Calcular X^T X
XtX = np.dot(X_matrix.T, X_matrix)

# Paso 6: Calcular la inversa de X^T X
XtX_inv = np.linalg.inv(XtX)

# Paso 7: Calcular X^T Y
XtY = np.dot(X_matrix.T, Y_matrix)

# Paso 8: Calcular beta
beta = np.dot(XtX_inv, XtY)

# Paso 9: Mostrar resultados
columnas = ['Intercepto', 'X1', 'X2', 'X3', 'X4']
beta_df = pd.DataFrame(beta, index=columnas, columns=['Beta'])
print("Coeficientes beta (regresión lineal múltiple):")
print(beta_df)

# Paso 10: Función para predecir Y
def predecir_Y():
    print("\n=== Predicción de Y ===")
    try:
        x1 = float(input("Ingresa el valor de X1: "))
        x2 = float(input("Ingresa el valor de X2: "))
        x3 = float(input("Ingresa el valor de X3: "))
        x4 = float(input("Ingresa el valor de X4: "))

        # Construimos el vector de entrada [1, X1, X2, X3, X4]
        x_nuevo = np.array([1, x1, x2, x3, x4])

        # Calculamos Ŷ = β0 + β1X1 + β2X2 + β3X3 + β4X4
        y_pred = np.dot(x_nuevo, beta).item()  # .item() convierte array de 1x1 a escalar

        print(f"Predicción del valor de Y: {y_pred:.2f}")

    except ValueError:
        print("Por favor, ingresa valores numéricos válidos.")

# Paso 11: Ejecutar predicción
predecir_Y()
```
