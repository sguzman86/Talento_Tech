```python
import pandas as pd

# Paso 1: Crear las listas de datos
variable_x = [1, 2, 2, 3, 4, 4, 5, 6]
variable_y = [2, 3, 4, 4, 4, 6, 5, 7]

# Paso 2: Crear el DataFrame con las variables X y Y
datos = {'X': variable_x, 'Y': variable_y}
tabla = pd.DataFrame(datos)

# Paso 3: Crear columnas auxiliares necesarias para las fórmulas
tabla['xy'] = tabla['X'] * tabla['Y']      # Producto x * y para cada fila
tabla['x^2'] = tabla['X'] ** 2             # Cuadrado de x para cada fila

# Paso 4: Calcular las sumatorias requeridas
n = len(tabla)                             # Número de datos (n)
sum_x = tabla['X'].sum()                   # ∑x
sum_y = tabla['Y'].sum()                   # ∑y
sum_xy = tabla['xy'].sum()                 # ∑xy
sum_x2 = tabla['x^2'].sum()                # ∑x²

# Paso 5: Aplicar la fórmula de la pendiente (m)
# m = (n∑xy - ∑x∑y) / (n∑x² - (∑x)²)
numerador_m = n * sum_xy - sum_x * sum_y
denominador_m = n * sum_x2 - sum_x ** 2
m = numerador_m / denominador_m

# Paso 6: Calcular la media de x y de y
x_promedio = sum_x / n
y_promedio = sum_y / n

# Paso 7: Aplicar la fórmula de la ordenada al origen (b)
# b = ȳ - m * x̄
b = y_promedio - m * x_promedio

# Paso 8: Mostrar resultados
print("Tabla de datos con columnas auxiliares:")
print(tabla)

print("\nSumatorias:")
print(f"n = {n}")
print(f"∑x = {sum_x}")
print(f"∑y = {sum_y}")
print(f"∑xy = {sum_xy}")
print(f"∑x² = {sum_x2}")

print("\nCálculos intermedios:")
print(f"Numerador de m = {numerador_m}")
print(f"Denominador de m = {denominador_m}")

print("\nResultados de la regresión lineal simple:")
print(f"Pendiente (m) = {round(m, 3)}")
print(f"Ordenada al origen (b) = {round(b, 3)}")

# Paso 9: Mostrar la ecuación de la recta
print(f"\nLa ecuación de la recta de regresión es: y = {round(m, 3)}x + {round(b, 3)}")
```

```python
import pandas as pd

# Paso 1: Crear las listas de datos
variable_x = [1, 2, 2, 3, 4, 4, 5, 6]
variable_y = [2, 3, 4, 4, 4, 6, 5, 7]

# Paso 2: Crear el DataFrame con las variables X y Y
datos = {'X': variable_x, 'Y': variable_y}
tabla = pd.DataFrame(datos)

# Paso 3: Crear columnas auxiliares necesarias
tabla['xy'] = tabla['X'] * tabla['Y']
tabla['x^2'] = tabla['X'] ** 2

# Paso 4: Calcular sumatorias
sumatorias = tabla.sum(numeric_only=True)

# Paso 5: Calcular pendiente y ordenada al origen
n = len(tabla)
sum_x = sumatorias['X']
sum_y = sumatorias['Y']
sum_xy = sumatorias['xy']
sum_x2 = sumatorias['x^2']

numerador_m = n * sum_xy - sum_x * sum_y
denominador_m = n * sum_x2 - sum_x ** 2
m = numerador_m / denominador_m

x_promedio = sum_x / n
y_promedio = sum_y / n
b = y_promedio - m * x_promedio

# Paso 6: Calcular Y_pred, error y error^2
tabla['Y_pred'] = m * tabla['X'] + b
tabla['Error'] = tabla['Y'] - tabla['Y_pred']
tabla['Error^2'] = tabla['Error'] ** 2

# Paso 7: Cálculo del MSE
MSE = tabla['Error^2'].mean()

# Paso 8: Calcular R^2
SST = ((tabla['Y'] - y_promedio) ** 2).sum()
SSE = tabla['Error^2'].sum()
R2 = 1 - (SSE / SST)

# Paso 9: Crear fila sumatoria y eliminar sumas en Y_pred y Error
sumatorias_actualizadas = tabla.sum(numeric_only=True)

# Convertir a diccionario para manipular
sum_dict = sumatorias_actualizadas.to_dict()

# Eliminar sumas innecesarias:
sum_dict['Y_pred'] = ''
sum_dict['Error'] = ''

# Crear fila sumatoria con las correcciones
fila_sumatoria = pd.DataFrame([sum_dict], index=['Sumatoria'])

# Paso 10: Concatenar tabla con la fila de sumatoria corregida
tabla_con_sumatoria = pd.concat([tabla, fila_sumatoria])

# Mostrar tabla final
print("Tabla con columnas auxiliares y análisis de regresión:")
print(tabla_con_sumatoria)

# Mostrar resultados
print("\nResultados de la regresión lineal simple:")
print(f"Pendiente (m) = {round(m, 3)}")
print(f"Ordenada al origen (b) = {round(b, 3)}")
print(f"Ecuación de la recta: y = {round(m, 3)}x + {round(b, 3)}")
print(f"MSE (Error Cuadrático Medio) = {round(MSE, 3)}")
print(f"Coeficiente de determinación (R²) = {round(R2, 3)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Paso 1: Crear las listas de datos
variable_x = [1, 2, 2, 3, 4, 4, 5, 6]
variable_y = [2, 3, 4, 4, 4, 6, 5, 7]

# Paso 2: Crear el DataFrame con las variables X y Y
datos = {'X': variable_x, 'Y': variable_y}
tabla = pd.DataFrame(datos)

# Paso 3: Crear columnas auxiliares necesarias
tabla['xy'] = tabla['X'] * tabla['Y']      # Producto X*Y
tabla['x^2'] = tabla['X'] ** 2             # X al cuadrado

# Paso 4: Calcular sumatorias necesarias
sumatorias = tabla.sum(numeric_only=True)

# Paso 5: Calcular parámetros de la recta (pendiente m y ordenada b)
n = len(tabla)
sum_x = sumatorias['X']
sum_y = sumatorias['Y']
sum_xy = sumatorias['xy']
sum_x2 = sumatorias['x^2']

numerador_m = n * sum_xy - sum_x * sum_y
denominador_m = n * sum_x2 - sum_x ** 2
m = numerador_m / denominador_m

x_promedio = sum_x / n
y_promedio = sum_y / n
b = y_promedio - m * x_promedio

# Paso 6: Calcular Y predictivo y errores
tabla['Y_pred'] = m * tabla['X'] + b               # Y estimado
tabla['Error'] = tabla['Y'] - tabla['Y_pred']      # Residuo
tabla['Error^2'] = tabla['Error'] ** 2             # Residuo al cuadrado

# Paso 7: Calcular sumatorias para las nuevas columnas
sum_y_pred = tabla['Y_pred'].sum()
sum_error = tabla['Error'].sum()
sum_error2 = tabla['Error^2'].sum()

# Agregar fila de sumatorias (subtotales relevantes)
fila_sumatoria = pd.DataFrame({
    'X': [sum_x],
    'Y': [sum_y],
    'xy': [sum_xy],
    'x^2': [sum_x2],
    'Y_pred': [sum_y_pred],
    'Error': [sum_error],
    'Error^2': [sum_error2]
}, index=['Sumatoria'])

tabla_con_sumatoria = pd.concat([tabla, fila_sumatoria])

# Paso 8: Calcular MSE y R²
MSE = sum_error2 / n
# Total Sum of Squares (SST)
sst = ((tabla['Y'] - y_promedio) ** 2).sum()
# Residual Sum of Squares (SSE)
sse = sum_error2
R2 = 1 - (sse / sst)

# Mostrar tabla con sumatorias
print("Tabla con columnas auxiliares y sumatorias relevantes:")
print(tabla_con_sumatoria)

# Mostrar resultados de la regresión y métricas
print("\nResultados de la regresión lineal simple:")
print(f"Pendiente (m) = {round(m, 3)}")
print(f"Ordenada al origen (b) = {round(b, 3)}")
print(f"Ecuación de la recta: y = {round(m, 3)}x + {round(b, 3)}")

print(f"\nMSE (Error Cuadrático Medio) = {round(MSE, 4)}")
print(f"Coeficiente de Determinación R² = {round(R2, 4)}")

# --- Graficar ---

plt.scatter(tabla['X'], tabla['Y'], color='blue', label='Datos reales (Y)')
plt.plot(tabla['X'], tabla['Y_pred'], color='red', label='Línea de regresión (Y_pred)')

for i in range(len(tabla)):
    plt.vlines(x=tabla['X'].iloc[i],
               ymin=tabla['Y'].iloc[i],
               ymax=tabla['Y_pred'].iloc[i],
               color='gray', linestyle='dashed', alpha=0.5)

plt.title('Regresión Lineal Simple con Evaluación')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()
plt.grid(True)
plt.show()
```
